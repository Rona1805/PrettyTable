```python
from prettytable import PrettyTable, ALL
import random

class1 = PrettyTable()
class1.field_names = ["ÖĞRENCİ NO", "ADI VE SOYADI","VİZE","FİNAL","GEÇME NOTU","BAŞARI DURUMU"] 
passersbyLeftovers = PrettyTable()
passersbyLeftovers.field_names = ["GEÇENLER", "KALANLAR"]

names = ["FÜSUN ARBAĞ", "İBRAHİM FUAT ÖRNEK", "EFE YAĞCI", "AHMET ÇAĞRI ÜNVEREN", "DENİZ ARMAĞAN KARDAŞ", "BERAAT ŞANLIER", "HANDE SINICI", "SERAP CEYHUN", "CENK YETER", "RÜYA SENEMTAŞI", "CÜNEYT GÜZELKÜÇÜK", 
         "BİLAL BARIŞ HELVACI", "AYLA GEYİK", "DURAN DEĞERLİYURT", "ENGİN ÖZATAY", "MÜMTAZ ALEMDAR", "HATİCE ÖZGE TOPÇUOĞLU", "MUSTAFA BUĞRA BÜYÜKKALAYCI", "MUSTAFA ABDULLAH ÖZÇOLAK", "GÜLİZ ALIŞIK"]
nameIndex = random.sample(range(0, len(names)), 10)
numbers = random.sample(range(1, 10000), 10)
midtermExam = random.sample(range(0,101),10)
finalExam = random.sample(range(0,101),10) 
passersBy = []
leftovers = []

for i in range(10):
  if (len(str(numbers[i])) != 4):
    studentNo = "21" + "0" * (4 - len(str(numbers[i]))) + str(numbers[i])
  else:
    studentNo = "21" + str(numbers[i])
    midtermGrade = float(midtermExam[i])
    finalGrade = float(finalExam[i])
    passingGrade = (midtermGrade * 0.4) + (finalGrade * 0.6)
    
    dictionary = { "Öğrenci No" : int(studentNo), "Adı Soyadı" : names[nameIndex[i]], "Vize Notu: " : int(midtermGrade), "Final Notu: " : int(finalGrade), "Geçme Notu" : int(passingGrade)}
    if passingGrade >= 50:
      dictionary2 = {"Başarı Durumu" : "GEÇTİ"}
      passersBy.append(dictionary["Adı Soyadı"])
    else:
      dictionary2 = {"Başarı Durumu" : "KALDI"}
      leftovers.append(dictionary["Adı Soyadı"])
    
    class1.add_row([dictionary["Öğrenci No"], dictionary["Adı Soyadı"], dictionary["Vize Notu: "], dictionary["Final Notu: "], dictionary["Geçme Notu"], dictionary2["Başarı Durumu"]])
if (len(passersBy) > len(leftovers)):
  condition = len(passersBy)
elif (len(passersBy) < len(leftovers)):
  condition = len(leftovers)
else:
  condition = len(leftovers)

for i in range(condition):
  if (i < len(passersBy) and i < len(leftovers)):
    passersbyLeftovers.add_row([passersBy[i], leftovers[i]])
  elif (i < len(passersBy) and i >= len(leftovers)):
    passersbyLeftovers.add_row([passersBy[i], "BOŞ"])
  elif (i < len(leftovers) and i >= len(passersBy)):
    passersbyLeftovers.add_row(["BOŞ", leftovers[i]])

class1.hrules = ALL
passersbyLeftovers.hrules = ALL
print(class1)
print("") 
print(passersbyLeftovers)
```
